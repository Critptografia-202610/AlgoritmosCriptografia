```python
def inverso(a,b):
    d, s, t, tabla = euclides_extendido_pasos(a, b)
    if s<0:
        s = s + b
    return s
```

```python
import string

ALFABETO = string.ascii_uppercase
M = 26  # módulo

def normalizar(texto: str) -> str:
    # Pasa todo a mayúsculas (y deja lo demás igual)
    return texto.upper()

def cifrar_afin(texto: str, a: int, b: int) -> str:
    if euclid(a, M) != 1:
        raise ValueError(f"'a' debe ser invertible mod {M}. gcd({a},{M})={euclid(a,M)}")
    texto = normalizar(texto)
    res = []
    for ch in texto:
        if ch in ALFABETO:
            m = ord(ch) - ord('A')          # A->0 ... Z->25
            c = (a*m + b) % M
            res.append(chr(c + ord('A')))
        else:
            res.append(ch)  # deja espacios, signos, etc.
    return "".join(res)

def descifrar_afin(texto_cifrado: str, a: int, b: int) -> str:
    if euclid(a, M) != 1:
        raise ValueError(f"'a' debe ser invertible mod {M}. gcd({a},{M})={euclid(a,M)}")
    a_inv = inverso(a, M)  # <-- TU FUNCIÓN
    texto_cifrado = normalizar(texto_cifrado)
    res = []
    for ch in texto_cifrado:
        if ch in ALFABETO:
            c = ord(ch) - ord('A')
            m = (a_inv * (c - b)) % M
            res.append(chr(m + ord('A')))
        else:
            res.append(ch)
    return "".join(res)
```

```python
a, b = 7, 10
mensaje = "FARO"
c = cifrar_afin(mensaje, a, b)
m = descifrar_afin(c, a, b)
print("Mensaje: ", mensaje)
print("Cifrado: ", c)
print("Descifrado:", m)
```
