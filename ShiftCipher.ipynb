```python
import string

ALFABETO = string.ascii_uppercase
M = 26

def normalizar(texto: str) -> str:
    return texto.upper()

def cifrar_shift(texto: str, k: int) -> str:
    texto = normalizar(texto)
    k = k % M
    res = []

    for ch in texto:
        if ch in ALFABETO:
            m = ord(ch) - ord('A')
            c = (m + k) % M
            res.append(chr(c + ord('A')))
        else:
            res.append(ch)

    return "".join(res)

def descifrar_shift(texto_cifrado: str, k: int) -> str:
    texto_cifrado = normalizar(texto_cifrado)
    k = k % M
    res = []

    for ch in texto_cifrado:
        if ch in ALFABETO:
            c = ord(ch) - ord('A')
            m = (c - k) % M
            res.append(chr(m + ord('A')))
        else:
            res.append(ch)

    return "".join(res)
```

```python
k = 10
mensaje = "FARO"

cifrado = cifrar_shift(mensaje, k)
descifrado = descifrar_shift(cifrado, k)

print("Mensaje:   ", mensaje)
print("Cifrado:   ", cifrado)
print("Descifrado:", descifrado)
```
